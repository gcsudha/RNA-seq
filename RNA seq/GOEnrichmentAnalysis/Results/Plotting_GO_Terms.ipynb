```r
library(ggplot2)
library(dplyr)
```

```r
combined = read.csv('C:/Users/sudha.gcupadhaya/Desktop/RNA-seq/GO_Enrichment_Analysis/Blast2go/Results/GO_Terms_DEGs/Top_20_combined.csv')
```

```r
unique(combined$Host)
```

```r
names(combined) = c('Gene_ID', 'GO_Name', 'Category','fdr', 'P', 'count', 'ref_count', 'Host', 'Percent')
```

```r
combined = combined[combined$Host=='Mint',]
```

```r
combined$GO_Name = with(combined, factor(GO_Name, levels= GO_Name[order(Category)]))
```

```r
options(repr.plot.width = 18, repr.plot.height = 8)
#tiff('C:/Users/sudha.gcupadhaya/Desktop/mint.tiff',width=18,height=8,units='in',res=300)
a = ggplot(combined, aes(x= GO_Name, y= count, fill = Category))+
           geom_bar(stat = 'identity')+
           theme_bw()+ 
       
        scale_y_continuous(limits = c(0,1250), breaks=c(0,250, 500, 750, 1000, 1250))+
    labs(title ='', x = 'GO Terms', y = '', fill = 'GO Category', color = 'black')+
    theme(axis.title.x= element_text(size =14, color ='black'),
    axis.title.y = element_text(size =14, color ='black'),
    axis.text.x = element_text(vjust =1, hjust = 1, size =12,angle =45,color = 'black'),
    axis.text.y = element_text(size =12, color = 'black'),
    plot.margin=unit(c(1,1,1,3),"cm"), # top, right, bottom, left  
    legend.text = element_text(size =12),
    legend.title = element_text(size =14))+
    
    theme(legend.position="bottom")+
    scale_fill_manual(labels = c('Biological Process', 'Cellular Component', 'Molecular Function'), 
                             values = c('indianred', 'orange', 'cornflowerblue'))
a
dev.off()
```

```r
options(repr.plot.width = 18, repr.plot.height = 20)
ggplot(combined, aes(x= GO_Name, y= count, fill = Category))+
           geom_col(position = 'stack')+
           theme_bw()+
    theme(axis.title.x= element_text(size =10, color ='black'),
    axis.title.y = element_text(size =10, color ='black'),
    axis.text.x = element_text(vjust =1, hjust = 1, size =10,angle =45,color = 'black'),
    axis.text.y = element_text(size =10, color = 'black'),
    plot.margin=unit(c(1,1,1,2),"cm"), # top, right, bottom, left  
    strip.text.x = element_text(size = 10),
    strip.text.y = element_text(size = 10),
    legend.text = element_text(size =10),
    legend.title = element_text(size =10))+
    facet_wrap(~Host, scale = 'free', ncol =1)
```
